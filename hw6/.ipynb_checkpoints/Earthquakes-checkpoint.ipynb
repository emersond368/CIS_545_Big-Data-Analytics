## Step 2.1 Earthquakes Only, truncate at 2 decimal places 

By converting to int * 100.  Then group and count.

```python
# TODO:
# Truncate to 2 decimal places
quake_df['near_lat'] = quake_df['Latitude']*100
quake_df['near_lat'] = quake_df['near_lat'].astype(int)
quake_df['near_lon'] = quake_df['Longitude']*100
quake_df['near_lon'] = quake_df['near_lon'].astype(int)
# Group by location and count quakes
#grouped = quake_df.groupby(['near_lat','near_lon']).count().reset_index
grouped = quake_df.groupby(['near_lat','near_lon']).count().reset_index()#.agg(['count'])
#grouped.sort_values(['Date'])
grouped = grouped.sort(['Date'],ascending=False)
grouped = grouped.rename(columns = {'Date':'quakes'})
grouped = grouped[['near_lat','near_lon','quakes']]
grouped
```

```python
test = grouped[grouped['quakes']==1]
test
```

# Step 2.2 Plot on Map

Color coding frequency within quantile: 1 -> yellow, 2 -> orange, 3 -> red, 4 -> white

```python
m = Basemap(projection='mill',llcrnrlat=-80,
            urcrnrlat=80, llcrnrlon=-180,urcrnrlon=180,
            lat_ts=20,resolution='c')

counts = (grouped['quakes']).tolist()
most = max(counts)

plt.figure(figsize=(12,10))
plt.title("Earthquakes")
m.fillcontinents(color='lightgreen',lake_color='aqua')
m.drawmapboundary(fill_color='aqua')
m.drawcoastlines()
m.drawcountries()

# Color-code from 1 --> 4 earthquakes in same area w/in 2 decimal places
colors = ['none', 'yellow', 'orange', 'red', 'white']

# TODO:
# Instead of plotting everything with the same color -- compute subsets
# of data with 1, 2, .., 4 different quakes, and plot with colors from
# the colors list above
grouped1 = grouped[grouped['quakes']==1]
longitudes = (grouped1['near_lon'] / 100).tolist()
latitudes = (grouped1['near_lat'] / 100).tolist()
x,y = m(longitudes,latitudes)
m.plot(x, y, "o", markersize = 3, color = 'yellow')

grouped2 = grouped[grouped['quakes']==2]
longitudes = (grouped2['near_lon'] / 100).tolist()
latitudes = (grouped2['near_lat'] / 100).tolist()
x,y = m(longitudes,latitudes)
m.plot(x, y, "o", markersize = 3, color = 'orange')

grouped3 = grouped[grouped['quakes']==3]
longitudes = (grouped3['near_lon'] / 100).tolist()
latitudes = (grouped3['near_lat'] / 100).tolist()
x,y = m(longitudes,latitudes)
m.plot(x, y, "o", markersize = 3, color = 'red')

grouped4 = grouped[grouped['quakes']==4]
longitudes = (grouped4['near_lon'] / 100).tolist()
latitudes = (grouped4['near_lat'] / 100).tolist()
x,y = m(longitudes,latitudes)
m.plot(x, y, "o", markersize = 3, color = 'white')

plt.show()
```

# Step 2.3 Find Locations with Multiple Quakes in a Month

```python
# TODO:
# compute DataFrame month_apart with pairs of earthquakes
# in the same (to 2 decimal places) location, which occur
# within a month of each other

from datetime import date
from dateutil.relativedelta import relativedelta
```

```python

result = pd.merge(quake_df, quake_df,  how='left', left_on=['near_lat','near_lon'], right_on = ['near_lat','near_lon'])
indices = []
for i in range(0,len(result)):
    if result.Date_x[i] != result.Date_y[i]:
        year = result['Date_x'].dt.year[i]
        month = result['Date_x'].dt.month[i]
        day = result['Date_x'].dt.day[i]
        test = date(year, month, day)
        year = result['Date_y'].dt.year[i]
        month = result['Date_y'].dt.month[i]
        day = result['Date_y'].dt.day[i]
        test2 = date(year, month, day)
        dist = abs(test2-test).total_seconds()/(3600*24)
        if dist < 32:
            indices.append(i)
start=len(result)
result2=result.copy(deep=True)
for i in range(0,len(indices)):
    result2 = result2.append(result[indices[i]:indices[i]+1])
stop = len(result2)
month_apart = result2[start:stop]        
month_apart2 = month_apart[['near_lat','near_lon']] 
month_apart2['quakes'] = 5
month_apart.head()
```
