```python
import numpy as np

def fft(time_data):
    return np.log10(np.absolute(np.fft.rfft(time_data, axis=1)[:,1:48]))
```

## Step 1.3

```python
# For each segment in data.keys(), take data[segment], transpose it, and call fft() on it.
# Take the result (all channels by all 47 frequencies) and convert it into a 1D array.
# This is a feature row in 2D array X.
# Set the class label for the same row in 1D array y, dependong on if the segment name 
# includes 'interictal' (this is a non-match) vs 'ictal' (this is a match)
features = np.empty((596,799))
iteration = 0
for labels in data:
    test = data[labels].T
    test2 = fft(test)
    test3 = np.reshape(test2, len(test2)*len(test2[0]))
    features[iteration,:] = test3
    iteration = iteration + 1
```

```python
y = data.keys()
new_y = []
for i in range(0,len(y)):
    new_y.append(y[i].split('_')[0])
new_y
```

```python
# Create X_train, X_test, y_train, y_test with
# random_state = 42 and test_size = 0.3
import sklearn.model_selection as ms
from sklearn.feature_extraction.text import TfidfTransformer
import numpy as np
X_train, X_test, y_train, y_test = ms.train_test_split(features,new_y, test_size=0.3)
y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
import sklearn.model_selection as ms
from sklearn.linear_model import LogisticRegression
import numpy as np
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

clfsvc = SVC(random_state=42)
clfsvc.fit(X_train, y_train)
predictsvc = clfsvc.predict(X_test)
```

```python
1 - len(np.where((predictsvc != y_test)== True)[0])/len(y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import roc_auc_score

# Results, as a list of dictionaries
classifier_results = []
rfc = {}
rfc['classifier']= 'RandomForest'
rfc['count'] = 31
clfrf = RandomForestClassifier(n_estimators=31, random_state=314)
clfrf.fit(X_train, y_train)
predictrf = clfrf.predict(X_test)
rfc['score']= 1 - len(np.where((predictrf != y_test)== True)[0])/len(y_test)

y_testb = np.array(y_test)
y_testb[y_testb=='interictal'] = 1
y_testb[y_testb =='ictal'] = 0
y_testb =y_testb.astype(np.int)

predictrfb = np.array(predictrf)
predictrfb[predictrfb=='interictal'] = 1
predictrfb[predictrfb=='ictal'] = 0
predictrfb = predictrfb.astype(np.int)
rfc['roc'] = roc_auc_score(y_testb, predictrfb)

classifier_results.append(rfc)

bdt = {}
bdt['classifier']= 'Bag-DecTree'
bdt['count'] = 31
cart = DecisionTreeClassifier(random_state=42)
clfbdt = BaggingClassifier(base_estimator=cart, n_estimators=31, random_state=314)
clfbdt.fit(X_train, y_train)
predictbdt = clfbdt.predict(X_test)
bdt['score']= 1- len(np.where((predictbdt != y_test)== True)[0])/len(y_test)

predictbdtb = np.array(predictbdt)
predictbdtb[predictbdtb=='interictal'] = 1
predictbdtb[predictbdtb=='ictal'] = 0
predictbdtb = predictbdtb.astype(np.int)
bdt['roc'] = roc_auc_score(y_testb, predictbdtb)

classifier_results.append(bdt)

blr1 = {}
blr1['classifier']= 'Bag-LogReg-L1'
blr1['count'] = 31
cart = LogisticRegression(penalty='l1',random_state=42,solver='liblinear')
clfblr1 = BaggingClassifier(base_estimator=cart, n_estimators=31, random_state=314)
clfblr1.fit(X_train, y_train)
predictblr1 = clfblr1.predict(X_test)
blr1['score'] = 1- len(np.where((predictblr1 != y_test)== True)[0])/len(y_test)

predictblr1b = np.array(predictblr1)
predictblr1b[predictblr1b=='interictal'] = 1
predictblr1b[predictblr1b=='ictal'] = 0
predictblr1b = predictblr1b.astype(np.int)
blr1['roc'] = roc_auc_score(y_testb, predictblr1b)

classifier_results.append(blr1)

blr2 = {}
blr2['classifier']= 'Bag-LogReg-L2'
blr2['count'] = 31
cart = LogisticRegression(penalty='l2',random_state=42,solver='liblinear')
clfblr2 = BaggingClassifier(base_estimator=cart, n_estimators=31, random_state=314)
clfblr2.fit(X_train, y_train)
predictblr2 = clfblr2.predict(X_test)
blr2['score'] =  1 - len(np.where((predictblr2 != y_test)== True)[0])/len(y_test)

predictblr2b = np.array(predictblr2)
predictblr2b[predictblr2b=='interictal'] = 1
predictblr2b[predictblr2b=='ictal'] = 0
predictblr2b = predictblr2b.astype(np.int)
blr2['roc'] = roc_auc_score(y_testb, predictblr2b)

classifier_results.append(blr2)

bsvc = {}
bsvc['classifier']= 'Bag-SVM'
bsvc['count'] = 31
cart = SVC(random_state=42)
clfsvc = BaggingClassifier(base_estimator=cart, n_estimators=31, random_state=314)
clfsvc.fit(X_train, y_train)
predictsvc = clfsvc.predict(X_test)
bsvc['score'] = 1 - len(np.where((predictsvc != y_test)== True)[0])/len(y_test)

predictsvcb = np.array(predictsvc)
predictsvcb[predictsvcb=='interictal'] = 1
predictsvcb[predictsvcb=='ictal'] = 0
predictsvcb = predictsvcb.astype(np.int)
bsvc['roc'] = roc_auc_score(y_testb, predictsvcb)

classifier_results.append(bsvc)


adt = {}
adt['classifier']= 'Boost-DecTree'
adt['count'] = 31
cart = DecisionTreeClassifier(random_state=42)
adaDT = AdaBoostClassifier(base_estimator=cart, n_estimators = 31, random_state=314)
adaDT.fit(X_train, y_train)
predictadaDT = adaDT.predict(X_test)
adt['score'] = 1 - len(np.where((predictadaDT != y_test)== True)[0])/len(y_test)

predictadaDTb = np.array(predictadaDT)
predictadaDTb[predictadaDTb=='interictal'] = 1
predictadaDTb[predictadaDTb=='ictal'] = 0
predictadaDTb = predictadaDTb.astype(np.int)

adt['roc'] = roc_auc_score(y_testb, predictadaDTb)

classifier_results.append(adt)

alr1 = {}
alr1['classifier']= 'Boost-LogReg-L1'
alr1['count'] = 31
cart = LogisticRegression(penalty='l1',random_state=42,solver='liblinear')
adalr1 = AdaBoostClassifier(base_estimator=cart, n_estimators = 31, random_state=314)
adalr1.fit(X_train, y_train)
predictadalr1 = adalr1.predict(X_test)
alr1['score'] = 1 - len(np.where((predictadalr1 != y_test)== True)[0])/len(y_test)

predictadalr1b = np.array(predictadalr1)
predictadalr1b[predictadalr1b=='interictal'] = 1
predictadalr1b[predictadalr1b=='ictal'] = 0
predictadalr1b = predictadalr1b.astype(np.int)

alr1['roc'] = roc_auc_score(y_testb, predictadalr1b)

classifier_results.append(alr1)

alr2 = {}
alr2['classifier']= 'Boost-LogReg-L2'
alr2['count'] = 31
cart = LogisticRegression(penalty='l2',random_state=42,solver='liblinear')
adalr2 = AdaBoostClassifier(base_estimator=cart, n_estimators = 31, random_state=314)
adalr2.fit(X_train, y_train)
predictadalr2 = adalr2.predict(X_test)
alr2['score'] = 1 - len(np.where((predictadalr2 != y_test)== True)[0])/len(y_test)

predictadalr2b = np.array(predictadalr2)
predictadalr2b[predictadalr2b=='interictal'] = 1
predictadalr2b[predictadalr2b=='ictal'] = 0
predictadalr2b = predictadalr2b.astype(np.int)

alr2['roc'] = roc_auc_score(y_testb, predictadalr2b)

classifier_results.append(alr2)

asvc = {}
asvc['classifier']= 'Boost-LogReg-L2'
asvc['count'] = 31
cart = SVC(random_state=42)
adasvc = AdaBoostClassifier(base_estimator=cart, algorithm = 'SAMME',n_estimators = 31, random_state=314)
adasvc.fit(X_train, y_train)
predictadasvc = adasvc.predict(X_test)
asvc['score'] = 1 - len(np.where((predictadasvc != y_test)== True)[0])/len(y_test)

predictadasvcb = np.array(predictadasvc)
predictadasvcb[predictadasvcb=='interictal'] = 1
predictadasvcb[predictadasvcb=='ictal'] = 0
predictadasvcb = predictadasvcb.astype(np.int)

asvc['roc'] = roc_auc_score(y_testb, predictadasvcb)

classifier_results.append(asvc)
```
